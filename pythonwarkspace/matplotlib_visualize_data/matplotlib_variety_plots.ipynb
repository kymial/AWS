### 바 차트 (막대 그래프)
x 축의 값이 분류가 가능한 카테고리 형태일 때 bar `bar()` , `barh()` 메서드로 막대 그래프를 표시 할 수 있다.  
주의! bar 메서드는 x축의 값이 첫 번째로 인수로 옴


````python
y = [2, 3, 1]
x = [1, 2, 3]
xlabel = ['가', '나', '다']

plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))

plt.show()
````

`xerr`, `yerr` 인수를 지정하면 에러 바를 추가할 수 있다.  
`alpha` 인수를 지정하면 바의 투명도를 지정할 수 있음 (0 : 완전 투명, 1 : 완전 불투명)


````python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_position = np.arange(4)
y_performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar h')
plt.barh(y_position, y_performance, xerr = error, alpha=0.4)
plt.yticks(y_position, people)
plt.xlabel('x 축')
plt.grid(True)
plt.show()
````

### 스템 플롯
이산확률이나 자기 상관관계를 표현한고자 할 때 사용하는 차트

### 파이 차트
각 분류가 전체에서 얼만큼 차지하고 있는지 파악하려 할때 사용하는 차트  
`pie()` 메서드로 표현할 수 있다.  
주의! 차트가 원형태를 유지할 수 있게 하려면 `axis('equal') 을 지정해야 한다.

````python
labels = ['개구리', '돼지', '개', '고양이']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

plt.title('파이 차트')
plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        colors=colors, 
        autopct='%1.1f%%', 
        shadow=True, 
        startangle=90
        )
plt.axis('equal')


plt.show()
        
````

### 스캐터 플롯
두 변수 간의 상관관계를 나타내고자 할 때 사용하는 차트  
`scatter()` 메서드로 표현할 수 있음  
점 하나의 위치는 x, y 값으로 표현됨



````python
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.title('스캐터 플롯')
plt.scatter(x, y)
plt.grid(True)
plt.show()
````

만약 x에 대하여 여러 변수를 한 차트로 비교하고자 할 때 점의 크기 혹은 색상으로 추가로 표현할 수 있다  
`scatter()` 메서드의 `c` 인수로 색상을 , `s` 인수로 크기를 지정할 수 있다.  


````python
n = 30
x = np.random.rand(n)
y1 = np.random.rand(n)
y2 = np.random.rand(n)
y3 = np.pi * (15 * np.random.rand(n)) ** 2
plt.title('버블 차트')
plt.scatter(x, y1, c=y2, s=y3)
plt.grid(True)
plt.show()
````

### Imshow
이미지 데이터 처럼 행과 열을 가진 2차원 데이터에 대하여 각 행렬 값을 색으로 표현하는 차트  
`imshow()` 메서드로 표현할 수 있음  


````python
# pip install scikit-learn
from sklearn.datasets import load_digits
````

````python
digits = load_digits()
x = digits.images[0]
x
````

````python
plt.title('mnist digits; 0')
plt.imshow(x, interpolation='nearest', cmap=plt.cm.bone_r)
plt.subplots_adjust(left=0.35, right=0.65, bottom=0.35, top=0.65)
plt.xticks([])
plt.yticks([])

plt.show()
````

`cmap` 인수로 데이터를 표현하는 칼라맵을 지정할 수 있음

````python
figure, axes = plt.subplots(1, 4, figsize=(12, 3), subplot_kw={'xticks': [], 'yticks': []})

axes[0].set_title('color map Blues')
axes[0].imshow(x, interpolation='nearest', cmap=plt.cm.Blues)

axes[1].set_title('color map Blues_r')
axes[1].imshow(x, interpolation='nearest', cmap=plt.cm.Blues_r)

axes[2].set_title('interpolation none')
axes[2].imshow(x, interpolation='none', cmap=plt.cm.bone_r)

axes[3].set_title('interpolation kaiser')
axes[3].imshow(x, interpolation='kaiser', cmap=plt.cm.bone_r)

plt.show()
````

### 컨투어 플롯
3차원 자료에서 x,y 입력과 z 출력이 존재할 때, z 표현을 명암이 아닌 선을 사용하는 차트  
`contour()` , `contourf()` 메서드를 사용  
`contour()` : 등고선만 표시  
`contourf()` : 색으로 표시  
입력 x, y에 대하여 1차원 배열을 그대로 사용할 수 없고, `meshgrid()` 명령으로 포인트 행렬을 만들어 사용해야 함


````python
def f(x, y) :
    return (1 - x /2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y **2)

n = 256
x =np.linspace(-3, 3, n)
y =np.linspace(-3, 3, n)
xx, yy = np.meshgrid(x, y)
zz = f(xx, yy)

plt.title('컨투어 플롯')
plt.contour(xx, yy, zz, colors='black')
plt.contourf(xx, yy, zz, alpha=.75, cmap='jet')
plt.grid(True)
plt.show()
````

### 3D 서피스 플롯
3차원 데이터를 입체적으로 표현하는 차트  
3d Axes 객체를 생성하여 사용함  
`plot_wireframe()` , `plot_surface()` 메서드로 표현  


````python
x= np.arange(-4, 4, 0.25)
y= np.arange(-4, 4, 0.25)
xx, yy = np.meshgrid(x, y)
rr = np.sqrt(xx**2 + yy**2)
zz = np.sin(rr)

axes = plt.axes(projection='3d') 

axes.set_title('3D 서피드 플롯')
axes.plot_surface(xx, yy, zz, rstride=1, cstride=1, cmap='hot')
plt.show()
````

### 삼각 그리드 생성
`Triangulation` 클래스로 삼각 그리드를 생성할 수 있다  
`Triangulation` 클래스의 생성자는 x, y, triangles 세 개의 인수를 받음  
`triangles`는 각 점의 기하학적 위상 정보  




````python
x = np.array([0, 1, 2])
y = np.array([0, np.sqrt(3), 0])
triangles = [[0, 1, 2]]
triangle = mtri.Triangulation(x, y, triangles)

plt.title('삼각 그리드')
plt.triplot(triangle, 'ko-')
plt.xlim(-0.1, 2.1)
plt.ylim(-0.1, 1.8)
plt.grid(True)

plt.show()
````

````python
x = np.asarray([0, 1, 2, 3, 4, 2])
y = np.asarray([0, np.sqrt(3), 0, np.sqrt(3), 0, 2 * np.sqrt(3)])
triangles = [[0, 1, 2], [2, 3, 4], [1, 2, 3], [1, 3, 5]]
triangle = mtri.Triangulation(x, y, triangles)

plt.title('여러 개 삼각형 그리기')
plt.triplot(triangle, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.grid(True)

plt.show()
````

### 삼각 그리드 세분화
`TriRefiner` , `UniformTriRefiner` 클래스로 그리드를 세분화 할 수 있음  
`refiner_triangulation()` : 단순히 삼각 그리드를 세분화  
`refine_field()` : 실제 함수에 최적화된 삼각 그리드 생성  

````python
refiner = mtri.UniformTriRefiner(triangle)
triangle2 = refiner.refine_triangulation(subdiv=2)

plt.title('삼각 그리드 세분화')
plt.triplot(triangle2, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.grid(True)

plt.show()
````

### 삼각 그리드 플롯
삼차원 그리드 상에 `tricontour`, `tricontourf`, `plot_trisur`, `plot_trisurf` 를 사용하여 2차원 등고선이나 3차원 표면을 그릴 수 있다.



````python
triangle3 = refiner.refine_triangulation(subdiv=5)
z = np.cos(1.5 * triangle3.x) * np.cos(1.5 * triangle3.y)

plt.title('삼각 그리드 등고선')
plt.tricontourf(triangle3, z, camp='jet')
plt.grid(True)
plt.show()
````

````python
triangle4 = refiner.refine_triangulation(subdiv=3)
z = np.cos(1.5 * triangle4.x) * np.cos(1.5 * triangle4.y)

axes = plt.axes(projection = '3d')
axes.set_title('삼각 3차원 그리드')
axes.plot_trisurf(triangle4.x, triangle4.y, z, cmap = 'jet' , linewidth=0.2)
axes.set_zlim(-1, 1)
axes.view_init(40, -40)

plt.show()
````
